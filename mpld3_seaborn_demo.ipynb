```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
x, y = np.random.normal(size=(2, 200))
color, size = np.random.random((2, 200))
```

```python
#create a figure and display it non-interactivly
fig, ax = plt.subplots()
ax.scatter(x, y, c=color, s=500 * size, alpha=0.3)
ax.grid(color='lightgray', alpha=0.7)
```

```python
#use mpld3 to display the figure interactively
import mpld3
mpld3.display(fig)
```

```python
#can also enable interactive plots for the entire notebook, untill it is disabled
mpld3.enable_notebook()
fig
```

```python
mpld3.disable_notebook()
fig
```
